### 2.4.1  Diseño de Capa de Concreto para Tubería de Acueducto Rural en Cali


En la construcción de un acueducto rural  cercano a la ciudad de Cali, se requiere calcular el volumen de concreto necesario para una capa de protección de una tubería de conducción de agua. La tubería tiene un grosor de 8 cm y un radio promedio de 14 cm, con una longitud de 500 metros (50000 cm).

cilindro.jpg

Sabiendo que el volumen de la tubería se calcula con la fórmula:
$$V_{tuberia}=\pi R^2l-\pi r^2 l, \tag{1}$$
donde $R$ es el radio externo de la tubería, $r$ es el radio interno de la tubería y $l$ es la longitud de la tubería.

Los ingenieros necesitan determinar el volumen exacto de concreto para la capa de protección, usaran una formula
para calcular el volumen donde se use el radio promedio de la tubería, el grosor de la tubería y longitud.

$$
V_{concreto} = 2 \pi \cdot \text{radio promedio} \cdot \text{longitud} \cdot \text{grosor}. \tag{2}
$$


Recuerda que el grosor es la diferencia entre el radio externo y el radio interno.

¿Es posible transformar algebraicamente la expresión (1) en (2)?
